```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib.gridspec import GridSpec
import matplotlib as mpl

# ---------------------------
# Style
# ---------------------------
mpl.rcParams.update({
    'font.family': 'sans-serif',
    'font.sans-serif': ['Arial'],
    'font.size': 10,
})

# ---------------------------
# Load data
# ---------------------------
df = pd.read_csv("downloads\datasets-list-organs-providers-cnt.csv")

df["cell_count"] = pd.to_numeric(df["cell_count"], errors="coerce")

# ---------------------------
# Define consortium groups FIRST
# ---------------------------
desired_order = ["HuBMAP", "HTAN", "SenNet", "GTEx", "SPARC"]

df["consortium"] = df["consortium"].apply(
    lambda x: x if x in desired_order else "Other"
)

# ---------------------------
# Aggregate AFTER relabeling
# ---------------------------
summary = (
    df.groupby(["organ", "consortium"], as_index=False)
      .agg(
          total_cells=("cell_count", "sum"),
          dataset_count=("dataset", "count")
      )
)

summary = summary.dropna(subset=["total_cells"])

# ---------------------------
# Color scaling
# ---------------------------
vmin = summary["total_cells"].min()
vmax = summary["total_cells"].max()
color_norm = LogNorm(vmin=max(vmin, 1), vmax=vmax)

# ---------------------------
# Size scaling
# ---------------------------
max_count = summary["dataset_count"].max()

dmax = 26.0
k = dmax / np.log1p(max_count)

summary["diam"] = k * np.log1p(summary["dataset_count"])
summary["s"] = summary["diam"] ** 2

# ---------------------------
# Positions
# ---------------------------
x_spacing = 1.2

organs = sorted(summary["organ"].unique())
consortia = desired_order + ["Other"]

x_map = {org: i * x_spacing for i, org in enumerate(organs)}
y_map = {c: i for i, c in enumerate(consortia)}

summary["x"] = summary["organ"].map(x_map)
summary["y"] = summary["consortium"].map(y_map)
summary
```

```python
fig_w = max(10, 0.6 * len(organs))
fig_h = 5

fig = plt.figure(figsize=(fig_w, fig_h))
gs = GridSpec(2, 1, height_ratios=[4, 1], hspace=0)

plt.subplots_adjust(top=0.98, bottom=0.4, hspace=0)

ax = fig.add_subplot(gs[0])
legend_ax = fig.add_subplot(gs[1])
legend_ax.axis("off")

sc = ax.scatter(
    summary["x"],
    summary["y"],
    s=summary["s"],
    c=summary["total_cells"],
    cmap="Blues",
    norm=color_norm,
    edgecolors="black",
    linewidths=0.4
)

# Axis formatting
ax.set_xticks([x_map[o] for o in organs])
ax.set_xticklabels(organs, rotation=45, ha="right")

ax.set_yticks([y_map[c] for c in consortia])
ax.set_yticklabels(consortia)

ax.set_ylim(-0.5, len(consortia)-0.5)
ax.grid(axis="y", linestyle=":", linewidth=0.6)

ax.set_ylabel("Consortium")
ax.set_xlabel("")

gs_bottom = GridSpec(
    1, 2,
    width_ratios=[1, 1],
    left=0.08,
    right=0.95,
    bottom=0.05,
    top=0.30,
    wspace=0.25
)

# ---- Colorbar ----
cax_parent = fig.add_subplot(gs_bottom[0])
cax_parent.axis("off")
cax = cax_parent.inset_axes([0.2, 0.45, 0.6, 0.06])

cbar = fig.colorbar(
    sc,
    cax=cax,
    orientation="horizontal"
)
cbar.set_label("Total cell count (log scale)")

# ---- Size legend ----
legend_ax2 = fig.add_subplot(gs_bottom[1])
legend_ax2.axis("off")

legend_counts = [1, 5, 10, 20]
legend_counts = [c for c in legend_counts if c <= max_count]

handles = []
for c in legend_counts:
    diam = k * np.log1p(c)
    handles.append(
        legend_ax2.scatter([], [], s=diam**2,
                           facecolors="none",
                           edgecolors="black")
    )

legend_ax2.legend(
    handles,
    [str(c) for c in legend_counts],
    title="# Datasets",
    loc="center",
    ncol=len(legend_counts),
    frameon=True,
    borderpad=1.2,
    columnspacing=1.5,
    handletextpad=1.0
)

# Resize top panel like original
# pos = ax.get_position()
# new_h = pos.height * 0.5
# ax.set_position([pos.x0, pos.y1 - new_h, pos.width, new_h])

plt.savefig(
    "output\organ_consortium_summary.svg",
    dpi=300,
    bbox_inches="tight"
)


plt.show()
```
